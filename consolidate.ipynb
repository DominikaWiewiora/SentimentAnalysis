## Import libraries

```python
import os
import pandas as pd
```

## consolidate csv files


```python
folder_path = "./Opinions"
df_list = []

try:
    all_items = os.listdir(folder_path)
    csv_files = [file for file in all_items if file.endswith('.csv')]
    

    for filename in csv_files:
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path, sep=',', encoding="utf-8") 
        df_list.append(df)
        
except FileNotFoundError:
    print(f"Error: folder '{folder_path}' not found. Check your path.")
    df_list = [] 

if df_list:
    opinions_df = pd.concat(df_list, ignore_index=True)

    output_path = "./consolidated_opinions.csv"
    opinions_df.to_csv(output_path, index=False, encoding="utf-8")
    
    print(f"combined {len(df_list)} files and saved to: {output_path}")
else:
    print("No CSV files found")
```
